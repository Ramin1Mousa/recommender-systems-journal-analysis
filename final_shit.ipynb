```python
entitiesToREmove = ('recommender systems',' ','recommandation system', 'user','collaboration','Collaborative')
```

```python
import collections
import pandas as pd
import matplotlib.pyplot as plt

data['KeyWords'] = data['KeyWords'].astype(str).str.lstrip('<').str.rstrip('+').replace('\n', '')
stopwords = set(line.strip() for line in STOPWORDS)
stopwords = stopwords.union(set(['mr','mrs','one','two','said']))

wordcount = {}
i=0
words=data['KeyWords']
for wordd in words:
    word = wordd.replace(".","")
    word = word.replace(":","")
    word=word.lower()
    word = word.replace("\"","")
    word = word.replace("!","")
    word = word.replace("\n","")
    word = word.replace("\t˜","")
    wordd = word.replace("*","")
    for word in wordd.split(','):
        if word not in STOPWORDS:
            if word not in wordcount:
                wordcount[word] = 1
            else:
                wordcount[word] += 1

# Print most common word
n_print = int(input("How many most common words to print: "))
print("\nOK. The {} most common words are as follows\n".format(n_print))
for x in entitiesToREmove:
    if x in wordcount.keys():
        del wordcount[x]
word_counter = collections.Counter(wordcount)

        
# Create a data frame of the most common words 
# Draw a bar chart
lst = word_counter.most_common(n_print)
df = pd.DataFrame(lst, columns = ['Word', 'Count'])
df.plot.bar(x='Word',y='Count')
```
